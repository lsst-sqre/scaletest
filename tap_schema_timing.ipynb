# Test TAP_SCHEMA query timing

```python
from lsst.rsp import get_tap_service, retrieve_query
import random
import time
```

```python
tap = get_tap_service("tap")
```

```python
n = random.choice(range(1, 10))
```

```python
queries = {
    f"SELECT TOP {n} * FROM TAP_SCHEMA.tables", 
    f"SELECT TOP {n} * FROM TAP_SCHEMA.columns ORDER BY column_name", 
    f"SELECT TOP {n} * FROM TAP_SCHEMA.schemas WHERE schema_name LIKE '%ivoa%'"
}
```

```python
# All these queries should take less than 1.5 seconds
for query in queries:
    results = tap.run_async(query)
    print(results[0])
```
