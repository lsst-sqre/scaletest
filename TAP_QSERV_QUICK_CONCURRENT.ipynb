Quick concurrent TAP search query on Qserv

```python
import random
from lsst.rsp import get_tap_service, retrieve_query
import pyvo
from concurrent.futures import ThreadPoolExecutor, as_completed
import time
```

```python
tap = get_tap_service("tap")
```

```python
queries = [
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 609788049252947289 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 611257134226610725 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 611257134226611508 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 609788049252941856 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 609788049252947054 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 609788049252947626 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 611257134226604952 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 609788049252941875 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 609788049252941924 LIMIT 100000001',
    'SELECT obj.objectId, obj.coord_ra, obj.coord_dec FROM dp1.Object AS obj WHERE obj.objectId = 611257134226611818 LIMIT 100000001',
]
```

```python
def run_query(query_num):
    start = time.time()
    try:
        query = random.choice(queries)
        results = tap.run_async(query)
        duration = time.time() - start
        return {
            "query_num": query_num,
            "status": "success",
            "duration": duration,
            "result": results
        }
    except Exception as e:
        duration = time.time() - start
        return {
            "query_num": query_num,
            "status": "failure",
            "duration": duration,
            "error": str(e)
        }
```

```python
with ThreadPoolExecutor(max_workers=12) as executor:
    futures = [executor.submit(run_query, i) for i in range(12)]

    for future in as_completed(futures):
        result = future.result()
        print(result['result'][0])
        assert result['status'] == 'success'
```
